```python
import pandas as pd
import numpy as np
```

```python
file_path = "data/25년_수확량 통계.xlsx"
xls = pd.ExcelFile(file_path)
sheet_names = xls.sheet_names
print(sheet_names)
```

```python
df = pd.read_excel(xls, sheet_name='Sheet1')
print(df.head())
```

```python
df
```

```python
df.groupby(['지역','CASE'])['수확량(kg/10a)'].mean()
```

```python
hs_df = df[df['지역'] == '화성'].reset_index(drop=True)
hs_df
```

```python
print('1.케이스별 수확량')
print(hs_df.groupby(['CASE'])['수확량(kg/10a)'].mean())
print('=============================================')
print('2.경작자별 케이스별 수확량')
print(hs_df.groupby(['경작자','CASE'])['수확량(kg/10a)'].mean())
```
