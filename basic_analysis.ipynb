```python
file_path = "data/25년_수확량 통계.csv"
# xls = pd.ExcelFile(file_path)
# sheet_names = xls.sheet_names
# print(sheet_names)
```

```python
# df = pd.read_excel(xls, sheet_name='Sheet1')
df= pd.read_csv(file_path)
print(df.head())
```

```python
print("--- [시작] 데이터 전처리(공백 및 타입 정리)를 시작합니다. ---")

# 2. [문자형] 컬럼 공백 제거 (앞/뒤 공백)
# 분석에 주로 사용되는 범주형 컬럼들
categorical_cols = [
    '지역', '코드번호', '경작자', '구분', 'CASE', '시비 처리', '작물', '품종'
]

for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip()

# 2-1. [문자형] '품종' 컬럼의 *중간* 공백도 모두 제거
if '품종' in df.columns:
    df['품종'] = df['품종'].str.replace(r'\s+', '', regex=True)

print("범주형(문자) 컬럼 공백 정리 완료.")

# 3. [숫자형] 문자열로 저장된 숫자 컬럼 정리
# (쉼표, 공백이 포함된 컬럼들)
numeric_object_cols = [
    '면적(m2)', '면적(평)', '수확량(kg)', '건중량(kg)',
    '평당 수확량', '건조 수확량(kg/10a)', '총 질소 살포량(kg/10a)'
]

for col in numeric_object_cols:
    if col in df.columns:
        # 1. 문자로 변환 (결측치는 'nan' 문자열이 됨)
        s = df[col].astype(str)
        # 2. 쉼표(,) 제거
        s = s.str.replace(',', '', regex=False)
        # 3. 모든 공백(앞/뒤/중간) 제거
        s = s.str.replace(r'\s+', '', regex=True)

        # 4. 숫자로 변환 (변환 안되는 값은 NaN 처리)
        df[col] = pd.to_numeric(s, errors='coerce')

print("숫자형(문자) 컬럼 숫자 타입 변환 완료.")

print("--- [완료] 데이터 전처리 완료. ---")

# 4. [확인] 전처리 후 데이터 정보 출력
print("\n--- [처리 후] 데이터 정보 (df.info()) ---")
df.info()
```

```python
# 4. [분석 대상] CASE 1과 2 데이터만 선택
df_target = df_25[df_25['CASE'].isin(['1', '2'])].copy()
print(f"CASE 1, 2 대상 데이터 {len(df_target)}개 필터링.")

# 5. [분석] 준수 여부 정의
# 사용자가 정의한 준수 기준
compliance_string = "밑거름 정량, 웃거름 변량"
df_target['준수 여부'] = (df_target['시비 처리'] == compliance_string)

# 6. [결과] 지역별, CASE별 준수율 계산
# 6.1. 그룹별로 '총 필지 수'와 '준수한 필지 수' 집계
compliance_stats = df_target.groupby(['지역', 'CASE']).agg(
    총_필지_수=('준수 여부', 'count'),  # 그룹 내 전체 행의 수
    준수_필지_수=('준수 여부', 'sum')   # '준수 여부'가 True(1)인 것들의 합
).reset_index()

# 6.2. 준수율 계산
compliance_stats['준수율 (%)'] = (compliance_stats['준수_필지_수'] / compliance_stats['총_필지_수']) * 100

print("\n--- [분석 결과] 지역별, CASE별 '시비 처리' 준수율 (CASE 1 & 2) ---")
print(compliance_stats.to_string(index=False, float_format="%.1f"))

# 7. (참고) '화성' 지역 CASE 2의 상세 내역 확인
print("\n--- (참고) '화성' 지역 CASE 2 상세 시비 처리 내역 ---")
hwaseong_case2_details = df_target[
    (df_target['지역'] == '화성') & (df_target['CASE'] == '2')
]['시비 처리'].value_counts()
print(hwaseong_case2_details)

# 7. (참고) '순창' 지역 CASE 2의 상세 내역 확인
print("\n--- (참고) '순창' 지역 CASE 2 상세 시비 처리 내역 ---")
sc_case2_details = df_target[
    (df_target['지역'] == '순창') & (df_target['CASE'] == '2')
]['시비 처리'].value_counts()
print(sc_case2_details)
```

```python
import pandas as pd
import plotly.express as px

# 1. 'df'가 메모리에 있다고 가정합니다.
# (안정성을 위해 df 로드 및 전처리 코드 포함)
try:
    if 'df' not in locals() or df.empty:
         print("DataFrame 'df'를 다시 로드합니다.")
         df = pd.read_csv("25년_수확량 통계.csv")
    else:
         print("기존 DataFrame 'df'를 사용합니다.")
except NameError:
    print("DataFrame 'df'를 새로 로드합니다.")
    df = pd.read_csv("25년_수확량 통계.csv")

# 2. 분석에 필요한 컬럼들 전처리
try:
    all_soil_components = [
        'soil_pH', 'soil_EC', 'soil_OM', 'soil_AVP', 'soil_AVSi', 'soil_K', 'soil_Ca', 'soil_Mg'
    ]
    for col in all_soil_components:
        df[col] = pd.to_numeric(df[col].astype(str).str.replace(',', '').str.strip(), errors='coerce')

    df['지역'] = df['지역'].str.strip()
    df['CASE'] = df['CASE'].astype(str).str.strip()
    df['코드번호'] = df['코드번호'].str.strip()
    df['시비 처리'] = df['시비 처리'].str.strip() # 시비 처리 컬럼 전처리

except Exception as e:
    print(f"데이터 전처리 중 오류 발생: {e}")
    raise

# 3. 데이터 필터링 (CASE 1, 2, 3 맥락 유지)
case_filter = ['1', '2', '3']
df_baseline = df[df['CASE'].isin(case_filter)].copy()

print(f"\n2025년 비교군(CASE 1, 2, 3) {len(df_baseline)}개 필터링.")

# 4. 데이터 형태 변환 (Melt) - '시비 처리'를 ID로 포함
melted_data_all = df_baseline.melt(
    id_vars=['지역', 'CASE', '코드번호', '시비 처리'],
    value_vars=all_soil_components,
    var_name='토양성분',
    value_name='값'
)

melted_data_all = melted_data_all.dropna(subset=['값'])
print(f"Melt된 전체 토양 성분 데이터 {len(melted_data_all)}개로 Box Plot 생성.")

# 5. Plotly를 사용한 Box Plot 시각화 (*** X축 및 색상 변경 ***)
fig = px.box(
    melted_data_all,
    x='시비 처리',              # *** [MODIFIED] X축: '시비 처리' (실제 처리) ***
    y='값',                      # Y축: 토양 성분 값
    color='시비 처리',           # *** [MODIFIED] 색상: '시비 처리' (실제 처리) ***
    facet_col='지역',            # 열 분리: 지역별
    facet_row='토양성분',        # 행 분리: 8개 토양 성분
    points='all',              # 모든 개별 데이터 포인트를 점으로 표시
    hover_data=['CASE'],       # [NEW] 마우스 오버 시 'CASE' (계획) 표시
    labels={
        "값": "성분 값",
        "CASE": "CASE (계획)",
        "지역": "지역",
        "토양성분": "토양 성분",
        "시비 처리": "시비 처리 (실제)"
    },
    title="지역별/*시비 처리별* 전체 토양 환경(Baseline) 비교 (8개 성분)"
)

# 6. 차트 레이아웃 가독성 높이기
fig.update_layout(
    height=2000,
    font=dict(family="Malgun Gothic, AppleGGothic, sans-serif"),
    showlegend=False
)
fig.update_yaxes(matches=None)
fig.update_annotations(align='right')
fig.update_xaxes(tickangle=-45) # X축 라벨 회전

fig.show()

image_filename = "토양성분.png"
fig.write_image(
        image_filename,
        width=1200,
        height=600,
        scale=2  # 2배 해상도
    )
```

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf

# 1. 'df'가 메모리에 있다고 가정합니다.
# (안정성을 위해 df 로드 및 전처리 코드 포함)
try:
    if 'df' not in locals() or df.empty:
         print("DataFrame 'df'를 다시 로드합니다.")
         df = pd.read_csv("25년_수확량 통계.csv")
    else:
         print("기존 DataFrame 'df'를 사용합니다.")
except NameError:
    print("DataFrame 'df'를 새로 로드합니다.")
    df = pd.read_csv("25년_수확량 통계.csv")

# 2. 분석에 필요한 컬럼들 전처리
try:
    # 수치형 1: 건조 수확량
    df['건조 수확량(kg/10a)'] = pd.to_numeric(df['건조 수확량(kg/10a)'].astype(str).str.replace(',', '').str.strip(), errors='coerce')

    # 수치형 2: 토양 성분 8가지
    all_soil_components = [
        'soil_pH', 'soil_EC', 'soil_OM', 'soil_AVP', 'soil_AVSi', 'soil_K', 'soil_Ca', 'soil_Mg'
    ]
    for col in all_soil_components:
        df[col] = pd.to_numeric(df[col].astype(str).str.replace(',', '').str.strip(), errors='coerce')

    # 범주형:
    df['지역'] = df['지역'].astype(str).str.strip()
    df['품종'] = df['품종'].astype(str).str.strip()

except Exception as e:
    print(f"데이터 전처리 중 오류 발생: {e}")
    raise

# 3. '화성' 지역 데이터 필터링 (품종은 '골든퀸3호'로 통일됨)
df_hwaseong = df[df['지역'] == '화성'].copy()
print(f"\n'화성' 지역 데이터 {len(df_hwaseong)}개 필터링.")

# 4. statsmodels 포뮬라를 위한 컬럼명 영문/단순화
rename_map = {
    '건조 수확량(kg/10a)': 'yield_kg_10a'
}
analysis_df_hwaseong = df_hwaseong.rename(columns=rename_map)


# --- [분석] '화성' 지역: 8개 토양 성분 각각의 수확량 설명력(R-squared) 비교 ---
print("\n--- [분석] '화성' 지역: 8개 토양 성분 각각의 수확량 설명력 비교 ---")
print("분석 모델: 각 성분별로 8개의 단순 선형 회귀 (수확량 ~ 토양성분)를 실행하고 R-squared 비교")

regression_results = []

for component in all_soil_components:
    # 1. 해당 성분과 수확량 데이터 (결측치 제거)
    model_cols = ['yield_kg_10a', component]
    model_data = analysis_df_hwaseong[model_cols].dropna()

    n_obs = len(model_data)

    # 2. 모델 학습 (데이터가 2개 초과일 때만)
    if n_obs > 2:
        try:
            formula = f'yield_kg_10a ~ {component}'
            model = smf.ols(formula, data=model_data).fit()

            # 3. 결과 추출
            r_squared = model.rsquared
            p_value = model.pvalues[component]

            regression_results.append({
                '토양 성분': component,
                'R-squared(설명력)': r_squared,
                'P-value(유의성)': p_value,
                '유효 데이터 수': n_obs
            })

        except Exception as e:
            regression_results.append({
                '토양 성분': component,
                'R-squared(설명력)': None,
                'P-value(유의성)': None,
                '유효 데이터 수': n_obs,
                '오류': str(e)
            })
    else:
        regression_results.append({
                '토양 성분': component,
                'R-squared(설명력)': None,
                'P-value(유의성)': None,
                '유효 데이터 수': n_obs
            })

# 4. 결과 취합 및 정렬
results_df = pd.DataFrame(regression_results)
results_df_sorted = results_df.sort_values(by='R-squared(설명력)', ascending=False)

print("\n--- [분석 결과] 토양 성분별 수확량 예측 설명력(R-squared) 순위 ---")
print(results_df_sorted.to_string(index=False, float_format="%.4f"))
```

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf

# 1. 'df'가 메모리에 있다고 가정합니다.
# (안정성을 위해 df 로드 및 전처리 코드 포함)
try:
    if 'df' not in locals() or df.empty:
         print("DataFrame 'df'를 다시 로드합니다.")
         df = pd.read_csv("data/25년_수확량 통계.csv")
    else:
         print("기존 DataFrame 'df'를 사용합니다.")
except NameError:
    print("DataFrame 'df'를 새로 로드합니다.")
    df = pd.read_csv("data/25년_수확량 통계.csv")

# 2. 분석에 필요한 컬럼들 전처리
try:
    # 수치형 1: 건조 수확량
    df['건조 수확량(kg/10a)'] = pd.to_numeric(df['건조 수확량(kg/10a)'].astype(str).str.replace(',', '').str.strip(), errors='coerce')

    # 수치형 2: 토양 성분 8가지
    all_soil_components = [
        'soil_pH', 'soil_EC', 'soil_OM', 'soil_AVP', 'soil_AVSi', 'soil_K', 'soil_Ca', 'soil_Mg'
    ]
    for col in all_soil_components:
        df[col] = pd.to_numeric(df[col].astype(str).str.replace(',', '').str.strip(), errors='coerce')

    # 범주형:
    categorical_cols = ['지역', '품종']
    for col in categorical_cols:
        df[col] = df[col].astype(str).str.strip()

except Exception as e:
    print(f"데이터 전처리 중 오류 발생: {e}")
    raise

# 3. [수정] '김제'와 '화성' 지역 데이터만 필터링 (전체 기간)
df_GH = df[df['지역'].isin(['김제', '화성'])].copy()
print(f"\n'김제'와 '화성' 지역 데이터 {len(df_GH)}개 필터링.")

# 4. statsmodels 포뮬라를 위한 컬럼명 영문/단순화
rename_map = {
    '건조 수확량(kg/10a)': 'yield_kg_10a',
    '지역': 'region',
    '품종': 'variety'
}
analysis_df = df_GH.rename(columns=rename_map)

# --- [모델 1] 토양 성분만 사용 (교란 변수 통제 X) ---
print("\n--- [모델 1] 수확량 ~ 8대 토양 성분 (김제+화성) ---")

# 1.1. 모델 1 데이터 준비 (결측치 제거)
model_1_cols = ['yield_kg_10a'] + all_soil_components
model_1_data = analysis_df[model_1_cols].dropna()
print(f"모델 1 (토양 성분만) 유효 데이터 수: {len(model_1_data)}")

# 1.2. 모델 1 포뮬라
formula_1 = 'yield_kg_10a ~ ' + ' + '.join(all_soil_components)

# 1.3. 모델 1 학습 및 요약
try:
    if len(model_1_data) > (len(all_soil_components) + 1):
        model_1 = smf.ols(formula_1, data=model_1_data).fit()
        print("\n--- [모델 1] 회귀분석 요약 결과 ---")
        print(model_1.summary())
    else:
        print("모델 1 회귀분석을 수행하기에 유효한 데이터가 부족합니다.")
except Exception as e:
    print(f"\n모델 1 회귀분석 중 오류 발생: {e}")

# --- [모델 2] 토양 성분 + 지역 + 품종 (교란 변수 통제 O) ---
print("\n\n--- [모델 2] 수확량 ~ 8대 토양 성분 + C(지역) + C(품종) (권장 모델) ---")

# 2.1. 범주형 변수 값 정리 (공백, 콤마, 특수문자 제거)
clean_cols = ['region', 'variety']
for col in clean_cols:
    analysis_df[col] = analysis_df[col].astype(str).str.replace(r'[ ,/-]', '_', regex=True)

# 2.2. 모델 2 데이터 준비 (결측치 제거)
model_2_cols = ['yield_kg_10a', 'region', 'variety'] + all_soil_components
model_2_data = analysis_df[model_2_cols].dropna()
print(f"모델 2 (토양+지역+품종) 유효 데이터 수: {len(model_2_data)}")

# 2.3. 모델 2 포뮬라
formula_2 = 'yield_kg_10a ~ ' + ' + '.join(all_soil_components) + ' + C(region) + C(variety)'

# 2.4. 모델 2 학습 및 요약
try:
    if len(model_2_data) > (len(all_soil_components) + analysis_df['region'].nunique() + analysis_df['variety'].nunique()):
        model_2 = smf.ols(formula_2, data=model_2_data).fit()
        print("\n--- [모델 2] 회귀분석 요약 결과 ---")
        print(model_2.summary())
    else:
        print("모델 2 회귀분석을 수행하기에 유효한 데이터가 부족합니다.")
except Exception as e:
    print(f"\n모델 2 회귀분석 중 오류 발생: {e}")
```

## 김제 및 화성 토양성분 데이터 비교

```python
import pandas as pd
import plotly.express as px
from scipy.stats import mannwhitneyu # 통계 검정 라이브러리

# 1. 'df'가 메모리에 있다고 가정합니다.
# (안정성을 위해 df 로드 및 전처리 코드 포함)
try:
    if 'df' not in locals() or df.empty:
         print("DataFrame 'df'를 다시 로드합니다.")
         df = pd.read_csv("data/25년_수확량 통계.csv")
    else:
         print("기존 DataFrame 'df'를 사용합니다.")
except NameError:
    print("DataFrame 'df'를 새로 로드합니다.")
    df = pd.read_csv("data/25년_수확량 통계.csv")

# 2. 분석에 필요한 컬럼들 전처리
all_soil_components = [
    'soil_pH', 'soil_EC', 'soil_OM', 'soil_AVP', 'soil_AVSi', 'soil_K', 'soil_Ca', 'soil_Mg'
]
try:
    # 8개 토양 성분 숫자형 변환
    for col in all_soil_components:
        df[col] = pd.to_numeric(df[col].astype(str).str.replace(',', '').str.strip(), errors='coerce')
    # '지역' 컬럼 공백 제거
    df['지역'] = df['지역'].astype(str).str.strip()
    df['코드번호'] = df['코드번호'].astype(str).str.strip()
except Exception as e:
    print(f"데이터 전처리 중 오류 발생: {e}")
    raise

# 3. '김제'와 '화성' 지역 데이터 필터링 (전체 기간)
df_regions = df[df['지역'].isin(['화성', '김제'])].copy()
print(f"'화성'과 '김제' 지역 전체 데이터 {len(df_regions)}개 필터링.")

# 4. [분석 1] 통계 분석 (Mann-Whitney U test)
gimje_data = df_regions[df_regions['지역'] == '김제']
hwaseong_data = df_regions[df_regions['지역'] == '화성']

stats_results = []
print("\n--- [분석 1] '김제' vs '화성' 토양 성분 비교 (중앙값 & P-value) ---")

for col in all_soil_components:
    # 각 지역의 유효한(NA가 아닌) 데이터 추출
    g_vals = gimje_data[col].dropna()
    h_vals = hwaseong_data[col].dropna()

    # 두 지역 모두 데이터가 있을 경우에만 테스트 실행
    if len(g_vals) > 0 and len(h_vals) > 0:
        g_median = g_vals.median()
        h_median = h_vals.median()
        # Mann-Whitney U test 실행 (두 집단이 다른지 검정)
        stat, p_value = mannwhitneyu(g_vals, h_vals, alternative='two-sided')

        stats_results.append({
            '토양성분': col,
            '김제 중앙값(Median)': g_median,
            '화성 중앙값(Median)': h_median,
            'P-value (유의성)': p_value
        })

# 결과 출력
stats_df = pd.DataFrame(stats_results)
print(stats_df.to_string(index=False, float_format="%.3f"))


# 5. [분석 2] 시각화 (Box Plot)
# 시각화를 위해 데이터 형태 변환 (Melt)
melted_data = df_regions.melt(
    id_vars=['지역', '코드번호'],
    value_vars=all_soil_components,
    var_name='토양성분',
    value_name='값'
)
melted_data = melted_data.dropna(subset=['값'])

fig = px.box(
    melted_data,
    x='지역',                  # X축: 지역 (김제 vs 화성)
    y='값',                      # Y축: 토양 성분 값
    color='지역',              # 색상: 지역별 구분
    facet_row='토양성분',        # 행 분리: 8개 토양 성분
    points='all',              # 모든 개별 데이터 포인트를 점으로 표시
    labels={
        "값": "성분 값",
        "지역": "지역"
    },
    title="'김제' vs '화성' 지역 8대 토양 성분 분포 비교"
)

# 레이아웃 설정
fig.update_layout(
    height=2000,
    font=dict(family="Malgun Gothic, AppleGGothic, sans-serif"),
    showlegend=False
)
fig.update_yaxes(matches=None) # 각 성분별 y축 독립
fig.update_annotations(align='right')

fig.show()

# 6. HTML 파일로 저장
html_filename = 'region_soil_comparison_boxplot.html'
fig.write_html(html_filename)

print(f"\n[분석 2] Box Plot 시각화 차트를 '{html_filename}' 파일로 저장했습니다.")
```
