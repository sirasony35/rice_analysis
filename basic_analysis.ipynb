```python
# df = pd.read_excel(xls, sheet_name='Sheet1')
df= pd.read_csv(file_path)
print(df.head())
```

```python
# 1. 대상 컬럼 리스트
# (이전 분석에서 'object' 타입이면서 숫자가 되어야 할 컬럼들)
cols_to_clean = [
    '면적(m2)', '면적(평)', '수확량(kg)', '건중량(kg)',
    '건조 수확량(kg/10a)', '총 질소 살포량(kg/10a)'
]

# 2. 반복문을 통해 각 컬럼을 순차적으로 클리닝
for col in cols_to_clean:
    if col in df.columns:

        # 3. 핵심 클리닝 및 변환 단계 (체이닝)
        #    (1) .astype(str) : 혹시 모를 숫자 타입을 문자열로 통일
        #    (2) .str.strip() : 값 앞뒤의 모든 공백(space) 제거
        #    (3) .str.replace(',', '') : 값 내부의 모든 쉼표(,) 제거
        cleaned_series = df[col].astype(str).str.strip().str.replace(',', '', regex=False)

        # 4. pd.to_numeric() : 깨끗해진 문자열을 숫자로 최종 변환
        #    (errors='coerce' 옵션: 변환 실패 시 NaN 값으로 처리)
        df[col] = pd.to_numeric(cleaned_series, errors='coerce')

    if 'CASE' in df.columns:
        df['CASE'] = df['CASE'].astype(str).str.strip()
        print("'CASE' 컬럼의 앞뒤 공백을 제거 했습니다.")

print("--- 변환 완료 ---")
df.info()
```

### 질소 시비량과 수확량 상관 분석

```python
import pandas as pd
import plotly.express as px

# 1. 'df'가 메모리에 있다고 가정합니다.
# (안정성을 위해 df 로드 및 전처리 코드 포함)
try:
    if 'df' not in locals() or df.empty:
         print("DataFrame 'df'를 다시 로드합니다.")
         df = pd.read_csv("25년_수확량 통계.csv")
    else:
         print("기존 DataFrame 'df'를 사용합니다.")
except NameError:
    print("DataFrame 'df'를 새로 로드합니다.")
    df = pd.read_csv("25년_수확량 통계.csv")

# 2. 분석에 필요한 컬럼들 전처리
try:
    # 수치형 열
    df['총 질소 살포량(kg/10a)'] = pd.to_numeric(df['총 질소 살포량(kg/10a)'].astype(str).str.replace(',', '').str.strip(), errors='coerce')
    df['건조 수확량(kg/10a)'] = pd.to_numeric(df['건조 수확량(kg/10a)'].astype(str).str.replace(',', '').str.strip(), errors='coerce')

    # 범주형 열
    df['지역'] = df['지역'].str.strip()
    df['품종'] = df['품종'].str.strip()
    df['시비 처리'] = df['시비 처리'].str.strip()
    df['코드번호'] = df['코드번호'].str.strip()

except Exception as e:
    print(f"데이터 전처리 중 오류 발생: {e}")
    raise

df_25 = df[df['년도'] == 2025].reset_index(drop=True)

# 3. 산점도를 위한 데이터 준비 (분석에 필요한 모든 컬럼에 결측치가 없는 행만 선택)
# 3.1. [NEW] '총 질소 살포량'이 0보다 큰 데이터만 필터링 (0 및 NA 제외)
df_filtered = df_25[df_25['총 질소 살포량(kg/10a)'] > 0].copy()
print(f"\n원본 {len(df_25)}개 중, 0보다 큰 질소 살포량 데이터 {len(df_filtered)}개 필터링.")

# 3.2. '총 질소 살포량' > 0 인 데이터 중에서 나머지 핵심 컬럼의 결측치(NA) 제거
key_columns = ['총 질소 살포량(kg/10a)', '건조 수확량(kg/10a)', '지역', '품종', '시비 처리', '코드번호']
plot_data_full = df_filtered.dropna(subset=key_columns).copy()
print(f"{len(df_filtered)}개 중, 수확량/품종 등 나머지 결측치 제거 후 최종 {len(plot_data_full)}개로 차트 생성.")

# 4. '지역'과 '품종'을 결합한 새 열 생성 (이전과 동일)
plot_data_full['지역_품종'] = plot_data_full['지역'] + ' (' + plot_data_full['품종'] + ')'

# 5. Plotly 시각화 (이전과 동일)
fig = px.scatter(
    plot_data_full,
    x='총 질소 살포량(kg/10a)',
    y='건조 수확량(kg/10a)',
    facet_col='지역_품종',
    color='시비 처리',
    hover_data=['코드번호'],
    labels={
        "총 질소 살포량(kg/10a)": "총 질소 살포량 (kg/10a)",
        "건조 수확량(kg/10a)": "건조 수확량 (kg/10a)",
        "지역_품종": "지역 (품종)",
        "시비 처리": "시비 처리 방식"
    },
    title="총 질소 살포량(0 제외)과 수확량의 관계 (지역/품종별 일렬 배치)"
)

# 6. 차트 레이아웃 가독성 높이기 (이전과 동일)
fig.update_layout(
    height=500,
    font=dict(family="Malgun Gothic, AppleGGothic, sans-serif")
)
fig.update_xaxes(title_text="총 질소 살포량 (kg/10a)")
fig.show()


image_filename = "총 질소 살포량 vs 수확량 상관관계.png"
fig.write_image(
        image_filename,
        width=1200,
        height=600,
        scale=2  # 2배 해상도
    )

# 4. 상관 계수 계산
corr_data = df_filtered[['총 질소 살포량(kg/10a)', '건조 수확량(kg/10a)']].dropna()

if len(corr_data) > 2:
    correlation_matrix = corr_data.corr(method='pearson')
    correlation = correlation_matrix.loc['총 질소 살포량(kg/10a)', '건조 수확량(kg/10a)']

    print(f"분석에 사용된 유효 데이터 수: {len(corr_data)}")
    print(f"피어슨 상관 계수: {correlation:.6f}")
else:
    print("상관분석을 수행하기에 유효한 데이터 수가 부족합니다.")
```

```python
import pandas as pd
import plotly.express as px
from scipy.stats import mannwhitneyu

# 1. 'df'가 메모리에 있다고 가정합니다.
# (안정성을 위해 df 로드 및 전처리 코드 포함)
try:
    if 'df' not in locals() or df.empty:
         print("DataFrame 'df'를 다시 로드합니다.")
         df = pd.read_csv("25년_수확량 통계.csv")
    else:
         print("기존 DataFrame 'df'를 사용합니다.")
except NameError:
    print("DataFrame 'df'를 새로 로드합니다.")
    df = pd.read_csv("25년_수확량 통계.csv")

# 2. 분석에 필요한 컬럼들 전처리
all_soil_components = [
    'soil_pH', 'soil_EC', 'soil_OM', 'soil_AVP', 'soil_AVSi', 'soil_K', 'soil_Ca', 'soil_Mg'
]
try:
    for col in all_soil_components:
        df[col] = pd.to_numeric(df[col].astype(str).str.replace(',', '').str.strip(), errors='coerce')
    df['지역'] = df['지역'].str.strip()
except Exception as e:
    print(f"데이터 전처리 중 오류 발생: {e}")
    raise

# 3. [분석 1] '김제'와 '화성' 지역 데이터 필터링 (전체 기간)
df_regions = df[df['지역'].isin(['화성', '김제'])].copy()
print(f"'화성'과 '김제' 지역 전체 데이터 {len(df_regions)}개 필터링.")

# 4. [분석 1] 통계 분석 (Mann-Whitney U test)
gimje_data = df_regions[df_regions['지역'] == '김제']
hwaseong_data = df_regions[df_regions['지역'] == '화성']

stats_results = []
print("\n--- [분석 1] '김제' vs '화성' 토양 성분 비교 (중앙값 & P-value) ---")

for col in all_soil_components:
    g_vals = gimje_data[col].dropna()
    h_vals = hwaseong_data[col].dropna()

    if len(g_vals) > 0 and len(h_vals) > 0:
        g_median = g_vals.median()
        h_median = h_vals.median()
        # 양측 검정 (두 집단이 다른가?)
        stat, p_value = mannwhitneyu(g_vals, h_vals, alternative='two-sided')

        stats_results.append({
            '토양성분': col,
            '김제 중앙값(Median)': g_median,
            '화성 중앙값(Median)': h_median,
            'P-value (유의성)': p_value
        })

# 결과 출력
stats_df = pd.DataFrame(stats_results)
print(stats_df.to_string(index=False, float_format="%.3f"))


# 5. [분석 2] 시각화 (Box Plot)
melted_data = df_regions.melt(
    id_vars=['지역', '코드번호'],
    value_vars=all_soil_components,
    var_name='토양성분',
    value_name='값'
)
melted_data = melted_data.dropna(subset=['값'])

fig = px.box(
    melted_data,
    x='지역',                  # X축: 지역 (김제 vs 화성)
    y='값',                      # Y축: 토양 성분 값
    color='지역',              # 색상: 지역별 구분
    facet_row='토양성분',        # 행 분리: 8개 토양 성분
    points='all',              # 모든 개별 데이터 포인트를 점으로 표시
    labels={
        "값": "성분 값",
        "지역": "지역"
    },
    title="'김제' vs '화성' 지역 8대 토양 성분 분포 비교"
)

fig.update_layout(
    height=2000,
    font=dict(family="Malgun Gothic, AppleGGothic, sans-serif"),
    showlegend=False
)
fig.update_yaxes(matches=None) # 각 성분별 y축 독립
fig.update_annotations(align='right')

fig.show()
```

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf

# 1. 'df'가 메모리에 있다고 가정합니다.
# (안정성을 위해 df 로드 및 전처리 코드 포함)
try:
    if 'df' not in locals() or df.empty:
         print("DataFrame 'df'를 다시 로드합니다.")
         df = pd.read_csv("25년_수확량 통계.csv")
    else:
         print("기존 DataFrame 'df'를 사용합니다.")
except NameError:
    print("DataFrame 'df'를 새로 로드합니다.")
    df = pd.read_csv("25년_수확량 통계.csv")

# 2. 분석에 필요한 컬럼들 전처리
try:
    # 수치형 1: 건조 수확량
    df['건조 수확량(kg/10a)'] = pd.to_numeric(df['건조 수확량(kg/10a)'].astype(str).str.replace(',', '').str.strip(), errors='coerce')

    # 수치형 2: 토양 성분 8가지
    all_soil_components = [
        'soil_pH', 'soil_EC', 'soil_OM', 'soil_AVP', 'soil_AVSi', 'soil_K', 'soil_Ca', 'soil_Mg'
    ]
    for col in all_soil_components:
        df[col] = pd.to_numeric(df[col].astype(str).str.replace(',', '').str.strip(), errors='coerce')

    # 범주형:
    df['지역'] = df['지역'].astype(str).str.strip()
    df['품종'] = df['품종'].astype(str).str.strip()

except Exception as e:
    print(f"데이터 전처리 중 오류 발생: {e}")
    raise

# 3. '화성' 지역 데이터 필터링 (품종은 '골든퀸3호'로 통일됨)
df_hwaseong = df[df['지역'] == '화성'].copy()
print(f"\n'화성' 지역 데이터 {len(df_hwaseong)}개 필터링.")

# 4. statsmodels 포뮬라를 위한 컬럼명 영문/단순화
rename_map = {
    '건조 수확량(kg/10a)': 'yield_kg_10a'
}
analysis_df_hwaseong = df_hwaseong.rename(columns=rename_map)


# --- [분석] '화성' 지역: 8개 토양 성분 각각의 수확량 설명력(R-squared) 비교 ---
print("\n--- [분석] '화성' 지역: 8개 토양 성분 각각의 수확량 설명력 비교 ---")
print("분석 모델: 각 성분별로 8개의 단순 선형 회귀 (수확량 ~ 토양성분)를 실행하고 R-squared 비교")

regression_results = []

for component in all_soil_components:
    # 1. 해당 성분과 수확량 데이터 (결측치 제거)
    model_cols = ['yield_kg_10a', component]
    model_data = analysis_df_hwaseong[model_cols].dropna()

    n_obs = len(model_data)

    # 2. 모델 학습 (데이터가 2개 초과일 때만)
    if n_obs > 2:
        try:
            formula = f'yield_kg_10a ~ {component}'
            model = smf.ols(formula, data=model_data).fit()

            # 3. 결과 추출
            r_squared = model.rsquared
            p_value = model.pvalues[component]

            regression_results.append({
                '토양 성분': component,
                'R-squared(설명력)': r_squared,
                'P-value(유의성)': p_value,
                '유효 데이터 수': n_obs
            })

        except Exception as e:
            regression_results.append({
                '토양 성분': component,
                'R-squared(설명력)': None,
                'P-value(유의성)': None,
                '유효 데이터 수': n_obs,
                '오류': str(e)
            })
    else:
        regression_results.append({
                '토양 성분': component,
                'R-squared(설명력)': None,
                'P-value(유의성)': None,
                '유효 데이터 수': n_obs
            })

# 4. 결과 취합 및 정렬
results_df = pd.DataFrame(regression_results)
results_df_sorted = results_df.sort_values(by='R-squared(설명력)', ascending=False)

print("\n--- [분석 결과] 토양 성분별 수확량 예측 설명력(R-squared) 순위 ---")
print(results_df_sorted.to_string(index=False, float_format="%.4f"))
```

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf

# 1. 'df'가 메모리에 있다고 가정합니다.
# (안정성을 위해 df 로드 및 전처리 코드 포함)
try:
    if 'df' not in locals() or df.empty:
         print("DataFrame 'df'를 다시 로드합니다.")
         df = pd.read_csv("25년_수확량 통계.csv")
    else:
         print("기존 DataFrame 'df'를 사용합니다.")
except NameError:
    print("DataFrame 'df'를 새로 로드합니다.")
    df = pd.read_csv("25년_수확량 통계.csv")

# 2. 분석에 필요한 컬럼들 전처리
try:
    # 수치형 1: 건조 수확량
    df['건조 수확량(kg/10a)'] = pd.to_numeric(df['건조 수확량(kg/10a)'].astype(str).str.replace(',', '').str.strip(), errors='coerce')

    # 수치형 2: 토양 성분 8가지
    all_soil_components = [
        'soil_pH', 'soil_EC', 'soil_OM', 'soil_AVP', 'soil_AVSi', 'soil_K', 'soil_Ca', 'soil_Mg'
    ]
    for col in all_soil_components:
        df[col] = pd.to_numeric(df[col].astype(str).str.replace(',', '').str.strip(), errors='coerce')

    # 범주형:
    categorical_cols = ['지역', '품종']
    for col in categorical_cols:
        df[col] = df[col].astype(str).str.strip()

except Exception as e:
    print(f"데이터 전처리 중 오류 발생: {e}")
    raise

# 3. [수정] '김제'와 '화성' 지역 데이터만 필터링 (전체 기간)
df_GH = df[df['지역'].isin(['김제', '화성'])].copy()
print(f"\n'김제'와 '화성' 지역 데이터 {len(df_GH)}개 필터링.")

# 4. statsmodels 포뮬라를 위한 컬럼명 영문/단순화
rename_map = {
    '건조 수확량(kg/10a)': 'yield_kg_10a',
    '지역': 'region',
    '품종': 'variety'
}
analysis_df = df_GH.rename(columns=rename_map)

# --- [모델 1] 토양 성분만 사용 (교란 변수 통제 X) ---
print("\n--- [모델 1] 수확량 ~ 8대 토양 성분 (김제+화성) ---")

# 1.1. 모델 1 데이터 준비 (결측치 제거)
model_1_cols = ['yield_kg_10a'] + all_soil_components
model_1_data = analysis_df[model_1_cols].dropna()
print(f"모델 1 (토양 성분만) 유효 데이터 수: {len(model_1_data)}")

# 1.2. 모델 1 포뮬라
formula_1 = 'yield_kg_10a ~ ' + ' + '.join(all_soil_components)

# 1.3. 모델 1 학습 및 요약
try:
    if len(model_1_data) > (len(all_soil_components) + 1):
        model_1 = smf.ols(formula_1, data=model_1_data).fit()
        print("\n--- [모델 1] 회귀분석 요약 결과 ---")
        print(model_1.summary())
    else:
        print("모델 1 회귀분석을 수행하기에 유효한 데이터가 부족합니다.")
except Exception as e:
    print(f"\n모델 1 회귀분석 중 오류 발생: {e}")

# --- [모델 2] 토양 성분 + 지역 + 품종 (교란 변수 통제 O) ---
print("\n\n--- [모델 2] 수확량 ~ 8대 토양 성분 + C(지역) + C(품종) (권장 모델) ---")

# 2.1. 범주형 변수 값 정리 (공백, 콤마, 특수문자 제거)
clean_cols = ['region', 'variety']
for col in clean_cols:
    analysis_df[col] = analysis_df[col].astype(str).str.replace(r'[ ,/-]', '_', regex=True)

# 2.2. 모델 2 데이터 준비 (결측치 제거)
model_2_cols = ['yield_kg_10a', 'region', 'variety'] + all_soil_components
model_2_data = analysis_df[model_2_cols].dropna()
print(f"모델 2 (토양+지역+품종) 유효 데이터 수: {len(model_2_data)}")

# 2.3. 모델 2 포뮬라
formula_2 = 'yield_kg_10a ~ ' + ' + '.join(all_soil_components) + ' + C(region) + C(variety)'

# 2.4. 모델 2 학습 및 요약
try:
    if len(model_2_data) > (len(all_soil_components) + analysis_df['region'].nunique() + analysis_df['variety'].nunique()):
        model_2 = smf.ols(formula_2, data=model_2_data).fit()
        print("\n--- [모델 2] 회귀분석 요약 결과 ---")
        print(model_2.summary())
    else:
        print("모델 2 회귀분석을 수행하기에 유효한 데이터가 부족합니다.")
except Exception as e:
    print(f"\n모델 2 회귀분석 중 오류 발생: {e}")
```

```python
import pandas as pd
import plotly.express as px

# 1. 'df'가 메모리에 있다고 가정합니다.
# (안정성을 위해 df 로드 및 전처리 코드 포함)
try:
    if 'df' not in locals() or df.empty:
         print("DataFrame 'df'를 다시 로드합니다.")
         df = pd.read_csv("25년_수확량 통계.csv")
    else:
         print("기존 DataFrame 'df'를 사용합니다.")
except NameError:
    print("DataFrame 'df'를 새로 로드합니다.")
    df = pd.read_csv("25년_수확량 통계.csv")

# 2. 분석에 필요한 컬럼들 전처리
try:
    # 수치형
    df['건조 수확량(kg/10a)'] = pd.to_numeric(df['건조 수확량(kg/10a)'].astype(str).str.replace(',', '').str.strip(), errors='coerce')
    df['soil_AVSi'] = pd.to_numeric(df['soil_AVSi'].astype(str).str.replace(',', '').str.strip(), errors='coerce')
    df['soil_Ca'] = pd.to_numeric(df['soil_Ca'].astype(str).str.replace(',', '').str.strip(), errors='coerce')

    # 범주형:
    df['지역'] = df['지역'].astype(str).str.strip()
    df['시비 처리'] = df['시비 처리'].astype(str).str.strip()
    df['코드번호'] = df['코드번호'].astype(str).str.strip()

except Exception as e:
    print(f"데이터 전처리 중 오류 발생: {e}")
    raise

# 3. '화성' 지역 데이터 필터링
df_hwaseong = df[df['지역'] == '화성'].copy()
print(f"\n'화성' 지역 데이터 {len(df_hwaseong)}개 필터링.")

# 4. 버블 차트를 위한 데이터 준비 (핵심 변수 4개 + 코드번호)
plot_data = df_hwaseong.dropna(subset=[
    '건조 수확량(kg/10a)',
    'soil_AVSi',  # X축
    'soil_Ca',    # Size
    '시비 처리',  # Color
    '코드번호'
])
print(f"버블 차트 분석에 유효한 데이터 {len(plot_data)}개 사용.")

# 5. *** [NEW] 상관 계수 계산 ***
corr_avsi = plot_data['건조 수확량(kg/10a)'].corr(plot_data['soil_AVSi'])
corr_ca = plot_data['건조 수확량(kg/10a)'].corr(plot_data['soil_Ca'])

print(f"\n[확인] Corr(수확량, 유효규산): {corr_avsi:.4f}")
print(f"[확인] Corr(수확량, 칼슘): {corr_ca:.4f}")

# 6. Plotly를 사용한 버블 차트 시각화 (이전과 동일)
fig = px.scatter(
    plot_data,
    x='soil_AVSi',
    y='건조 수확량(kg/10a)',
    color='시비 처리',
    size='soil_Ca',
    size_max=30,
    trendline='ols',
    trendline_scope='overall',
    hover_data=['코드번호', 'soil_Ca'],
    labels={
        "건조 수확량(kg/10a)": "건조 수확량 (kg/10a)",
        "soil_AVSi": "Soil 유효규산 (soil_AVSi)",
        "soil_Ca": "Soil 칼슘 (soil_Ca)",
        "시비 처리": "시비 처리 방식"
    },
    title="화성 지역: 수확량 vs 유효규산(X), 칼슘(Size), 시비 처리(Color) (단일 추세선)"
)

# 7. *** [NEW] 차트에 텍스트(상관 계수) 추가 ***
fig.add_annotation(
    text=f"Corr(수확량, 유효규산): <b>{corr_avsi:.3f}</b>", # R(Y, X)
    align='left',
    showarrow=False,
    xref='paper',  # 'paper'는 차트 영역 기준
    yref='paper',
    x=0.05,        # 왼쪽에서 5%
    y=0.98         # 위에서 2%
)
fig.add_annotation(
    text=f"Corr(수확량, 칼슘): <b>{corr_ca:.3f}</b>", # R(Y, Size)
    align='left',
    showarrow=False,
    xref='paper',
    yref='paper',
    x=0.05,
    y=0.93         # 첫번째 텍스트 아래
)

# 8. 차트 레이아웃 가독성 높이기 (생략)
fig.update_layout(
    height=600,
    font=dict(family="Malgun Gothic, AppleGGothic, sans-serif")
)

fig.show()

# 9. HTML 파일로 저장
html_filename = 'hwaseong_avsi_ca_vs_yield.html'
fig.write_html(html_filename)

print(f"\nPlotly 버블 차트를 '{html_filename}' 파일로 저장했습니다.")
print("HTML 파일을 열어 수확량(Y), 규산(X), 칼슘(크기), 시비처리(색)의 관계를 확인하세요.")
```
