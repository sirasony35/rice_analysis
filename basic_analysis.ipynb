```python
import pandas as pd
import numpy as np
```

```python
file_path = "data/25년_수확량 통계.csv"
# xls = pd.ExcelFile(file_path)
# sheet_names = xls.sheet_names
# print(sheet_names)
```

```python
# df = pd.read_excel(xls, sheet_name='Sheet1')
df= pd.read_csv(file_path)
print(df.head())
```

```python
# 1. 대상 컬럼 리스트
# (이전 분석에서 'object' 타입이면서 숫자가 되어야 할 컬럼들)
cols_to_clean = [
    '면적(m2)', '면적(평)', '수확량(kg)', '건중량(kg)',
    '건조 수확량(kg/10a)', '총 질소 살포량(kg/10a)'
]

# 2. 반복문을 통해 각 컬럼을 순차적으로 클리닝
for col in cols_to_clean:
    if col in df.columns:

        # 3. 핵심 클리닝 및 변환 단계 (체이닝)
        #    (1) .astype(str) : 혹시 모를 숫자 타입을 문자열로 통일
        #    (2) .str.strip() : 값 앞뒤의 모든 공백(space) 제거
        #    (3) .str.replace(',', '') : 값 내부의 모든 쉼표(,) 제거
        cleaned_series = df[col].astype(str).str.strip().str.replace(',', '', regex=False)

        # 4. pd.to_numeric() : 깨끗해진 문자열을 숫자로 최종 변환
        #    (errors='coerce' 옵션: 변환 실패 시 NaN 값으로 처리)
        df[col] = pd.to_numeric(cleaned_series, errors='coerce')

    if 'CASE' in df.columns:
        df['CASE'] = df['CASE'].astype(str).str.strip()
        print("'CASE' 컬럼의 앞뒤 공백을 제거했습니다.")

print("--- 변환 완료 ---")
df.info()
```

```python
df_25 = df[df['년도']==2025]
df_25
```

```python
graph_yield = df_25.groupby(['지역','CASE'])['건조 수확량(kg/10a)'].mean().reset_index()
graph_yield.rename(columns={'건조 수확량(kg/10a)': '평균 건조 수확량(kg/10a)'}, inplace=True)
graph_yield
```

```python
import pandas as pd
import plotly.express as px
```

```python
# --- 3. Plotly 막대 그래프 생성 코드 ---

fig = px.bar(
    graph_yield,  # 사용할 데이터프레임
    x='CASE',         # X축: CASE
    y='평균 건조 수확량(kg/10a)', # Y축: 평균 수확량
    color='CASE',     # 색상: CASE별로 다르게
    facet_col='지역', # 차트 분리: '지역'별로 차트를 나누어 그림
    title='2025년 지역별/CASE별 평균 건조 수확량(kg/10a)', # 차트 제목
    text_auto=True,    # 막대 위에 수확량 값 표시
    category_orders={"CASE": ['1', '2', '3', '4']}
)

# Y축 범위 0부터 시작하도록 설정 (선택 사항)
fig.update_yaxes(range=[0, None])

# 제목 가운데 정렬
fig.update_layout(title_x=0.5)
fig.show()
```
