```python
# df = pd.read_excel(xls, sheet_name='Sheet1')
df= pd.read_csv(file_path)
print(df.head())
```

```python
# 1. 대상 컬럼 리스트
# (이전 분석에서 'object' 타입이면서 숫자가 되어야 할 컬럼들)
cols_to_clean = [
    '면적(m2)', '면적(평)', '수확량(kg)', '건중량(kg)',
    '건조 수확량(kg/10a)', '총 질소 살포량(kg/10a)'
]

# 2. 반복문을 통해 각 컬럼을 순차적으로 클리닝
for col in cols_to_clean:
    if col in df.columns:

        # 3. 핵심 클리닝 및 변환 단계 (체이닝)
        #    (1) .astype(str) : 혹시 모를 숫자 타입을 문자열로 통일
        #    (2) .str.strip() : 값 앞뒤의 모든 공백(space) 제거
        #    (3) .str.replace(',', '') : 값 내부의 모든 쉼표(,) 제거
        cleaned_series = df[col].astype(str).str.strip().str.replace(',', '', regex=False)

        # 4. pd.to_numeric() : 깨끗해진 문자열을 숫자로 최종 변환
        #    (errors='coerce' 옵션: 변환 실패 시 NaN 값으로 처리)
        df[col] = pd.to_numeric(cleaned_series, errors='coerce')

    if 'CASE' in df.columns:
        df['CASE'] = df['CASE'].astype(str).str.strip()
        print("'CASE' 컬럼의 앞뒤 공백을 제거 했습니다.")

print("--- 변환 완료 ---")
df.info()
```

### 질소 시비량과 수확량 상관 분석

```python
import pandas as pd
import plotly.express as px

# 1. 'df'가 메모리에 있다고 가정합니다.
# (안정성을 위해 df 로드 및 전처리 코드 포함)
try:
    if 'df' not in locals() or df.empty:
         print("DataFrame 'df'를 다시 로드합니다.")
         df = pd.read_csv("25년_수확량 통계.csv")
    else:
         print("기존 DataFrame 'df'를 사용합니다.")
except NameError:
    print("DataFrame 'df'를 새로 로드합니다.")
    df = pd.read_csv("25년_수확량 통계.csv")

# 2. 분석에 필요한 컬럼들 전처리
try:
    # 수치형 열
    df['총 질소 살포량(kg/10a)'] = pd.to_numeric(df['총 질소 살포량(kg/10a)'].astype(str).str.replace(',', '').str.strip(), errors='coerce')
    df['건조 수확량(kg/10a)'] = pd.to_numeric(df['건조 수확량(kg/10a)'].astype(str).str.replace(',', '').str.strip(), errors='coerce')

    # 범주형 열
    df['지역'] = df['지역'].str.strip()
    df['품종'] = df['품종'].str.strip()
    df['시비 처리'] = df['시비 처리'].str.strip()
    df['코드번호'] = df['코드번호'].str.strip()

except Exception as e:
    print(f"데이터 전처리 중 오류 발생: {e}")
    raise

df_25 = df[df['년도'] == 2025].reset_index(drop=True)

# 3. 산점도를 위한 데이터 준비 (분석에 필요한 모든 컬럼에 결측치가 없는 행만 선택)
# 3.1. [NEW] '총 질소 살포량'이 0보다 큰 데이터만 필터링 (0 및 NA 제외)
df_filtered = df_25[df_25['총 질소 살포량(kg/10a)'] > 0].copy()
print(f"\n원본 {len(df_25)}개 중, 0보다 큰 질소 살포량 데이터 {len(df_filtered)}개 필터링.")

# 3.2. '총 질소 살포량' > 0 인 데이터 중에서 나머지 핵심 컬럼의 결측치(NA) 제거
key_columns = ['총 질소 살포량(kg/10a)', '건조 수확량(kg/10a)', '지역', '품종', '시비 처리', '코드번호']
plot_data_full = df_filtered.dropna(subset=key_columns).copy()
print(f"{len(df_filtered)}개 중, 수확량/품종 등 나머지 결측치 제거 후 최종 {len(plot_data_full)}개로 차트 생성.")

# 4. '지역'과 '품종'을 결합한 새 열 생성 (이전과 동일)
plot_data_full['지역_품종'] = plot_data_full['지역'] + ' (' + plot_data_full['품종'] + ')'

# 5. Plotly 시각화 (이전과 동일)
fig = px.scatter(
    plot_data_full,
    x='총 질소 살포량(kg/10a)',
    y='건조 수확량(kg/10a)',
    facet_col='지역_품종',
    color='시비 처리',
    hover_data=['코드번호'],
    labels={
        "총 질소 살포량(kg/10a)": "총 질소 살포량 (kg/10a)",
        "건조 수확량(kg/10a)": "건조 수확량 (kg/10a)",
        "지역_품종": "지역 (품종)",
        "시비 처리": "시비 처리 방식"
    },
    title="총 질소 살포량(0 제외)과 수확량의 관계 (지역/품종별 일렬 배치)"
)

# 6. 차트 레이아웃 가독성 높이기 (이전과 동일)
fig.update_layout(
    height=500,
    font=dict(family="Malgun Gothic, AppleGGothic, sans-serif")
)
fig.update_xaxes(title_text="총 질소 살포량 (kg/10a)")
fig.show()


image_filename = "총 질소 살포량 vs 수확량 상관관계.png"
fig.write_image(
        image_filename,
        width=1200,
        height=600,
        scale=2  # 2배 해상도
    )

# 4. 상관 계수 계산
corr_data = df_filtered[['총 질소 살포량(kg/10a)', '건조 수확량(kg/10a)']].dropna()

if len(corr_data) > 2:
    correlation_matrix = corr_data.corr(method='pearson')
    correlation = correlation_matrix.loc['총 질소 살포량(kg/10a)', '건조 수확량(kg/10a)']

    print(f"분석에 사용된 유효 데이터 수: {len(corr_data)}")
    print(f"피어슨 상관 계수: {correlation:.6f}")
else:
    print("상관분석을 수행하기에 유효한 데이터 수가 부족합니다.")
```
